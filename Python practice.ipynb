```python
x = [1,2,3]
#for loop in python
for i in x:
    if (i %2 == 0):
        print("It's even")
    else :
        print("It's odd")
```

```python
x = [1,2,3]
count = 0
#for loop in python
for i in x:  ## {In for loop , we give a list or array} 
    if (i %2 == 0):
        print("It's even")
    else :
        print("It's odd")
        count = count + 1
print("Number of odd numbers:",count)
```

```python
## Break and continue
## Break - comes out of all iterations
## Continue - comes out of that iteration
x = [1,2,3,7]
oddcount = 0
evencount = 0
#for loop in python
for i in x:  ## {In for loop , we give a list or array} 
    if  i >= 7 : ## If we don't want 7
        break
    elif (i %2 == 0):
        print("It's even")
        evencount = evencount + 1
    else:
            print("It's odd")
            oddcount = oddcount + 1
print("Number of odd numbers:",oddcount)
print("Number of even numbers:",evencount)
```

```python
x = [1,2,3,7]
oddcount = 0
evencount = 0
#for loop in python
for i in x:  ## {In for loop , we give a list or array} 
    if  i == 7 : ## If we don't want 7
        break
        if (i %2 == 0):
            print("It's even")
            evencount = evencount + 1
    else:
            print("It's odd")
            oddcount = oddcount + 1
print("Number of odd numbers:",oddcount)
print("Number of even numbers:",evencount)
```
