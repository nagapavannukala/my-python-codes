### 5. Build a linear regression model to predict the dependent variable Temp, using CO2, N2O, CFC.12, and Aerosols as independent variables. 

### Find the RMSE, SSE and MAPE for the above Model. What is the model R2 (the "Multiple R-squared" value & Adjusted R2).

```python
import statsmodels.api as sm
```

```python
x=sm.add_constant(data[["CO2", "N2O", "CFC-12", "Aerosols" ]])
```

```python
y=data.Temp
```

```python
lm = sm.OLS( y, x ).fit()
```

### Adjusted_R_square=0.591

```python
lmpredict = lm.predict(x)
```

```python
lmpredict[0:4]
```

```python
from statsmodels.tools.eval_measures import rmse
```

```python
SSE=sum((y-lmpredict)**2)
```

```python
3**2
```

```python
def mape_vectorized_v2(a, b): 
    mask = a != 0
    return (np.fabs(a - b)/a)[mask].mean()  ## one more way to find MAPE
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y, lmpredict)
```

### 8. What is the Hypothesis test in case of a Linear Regression?

### Ans:

#### Null Hypothesis: There is no significant linear relationship between the independent variable x and the dependent variable y (B1=B2=B3....=0)

#### Alternative Hypothesis: There is a no significant linear relationship between the independent variable x and the dependent variable y (B1 !=0). Atleast one variable is significant.

### 9. What do the P values represent? Do we look for low or High P values? Why?

### Ans:

#### A p value is used in hypothesis testing to help you support or reject the null hypothesis. The p value is the evidence against a null hypothesis. 
#### The smaller the p-value, the strong the evidence that you should reject the null hypothesis, so we normally look for smaller p-value.

### 10. What is an F Test in Linear Regression? Discuss.

### Ans:

#### The F-test of overall significance indicates whether the  linear regression model provides a better fit to the data than a model that contains no independent variables. 

#### In general, an F-test in regression compares the fits of different linear models. Unlike t-tests that can assess only one regression coefficient at a time, the F-test can assess multiple coefficients simultaneously.

#### If the P value for the F-test of overall significance test is less than your significance level, we can reject the null-hypothesis and conclude that your model provides a better fit than the intercept-only model.
