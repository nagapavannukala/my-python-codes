## Importing movies.dat file

```python
file = open('movies.dat')
lst = []
for line in file:
    lst += [line.split()]
```

```python
df = pd.DataFrame(lst)
df.columns = ['Column1','Column2','Column3','Column4','Column5','Column6','Column7','Column8','Column9','Column10',
              'Column11','Column12','Column13','Column14','Column15','Column16']
```

## Here we saw the shape of the Dataframe and created labels of the columns , so that it would be easy for merging into one dataframe

```python
df['ColumnA'] = df[df.columns[0:]].apply(lambda x: ' '.join(x.dropna().astype(str)),axis=1)
df
```

```python
mov = []
mov = df['ColumnA']
mov
```
