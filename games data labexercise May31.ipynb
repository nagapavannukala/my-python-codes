```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
data = pd.read_csv('games.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.dropna(thresh=20).info()
```

```python
nomiss = data.dropna(thresh=20)
nomiss.head()
```

```python
sns.pairplot(nomiss[['playingtime','minage','average_rating']])
plt.show()
```

### The above plot shows the relationship between Playing time , Min age, Average rating. Observing the upper triangle,
### We can see that there is o considerable correlation between each of them.(As the maximum datapoints tend to form either a horizontal or a vertical line)

```python
sns.jointplot(nomiss['average_rating'],nomiss['total_owners'], kind="kde")
plt.show()
```

### In The above graph, the individual distribution plots of average rating is not looking much like a normal distribution.
### The distribution plot of total_owners appeaars to have a huge right skew.

```python
sns.barplot(nomiss['minage'],nomiss['playingtime'])
plt.show()
```

## The above plot infers that at 14 years , the Peak playing time is observed.

```python
sns.lmplot(x="users_rated",y="total_owners",data = nomiss)
plt.show()
```

### There is a strong correlation between users_rated and total_owners. As the users who rated increase , it says that there is an increase in the number of Total owners.

```python
sns.pairplot(nomiss[['total_traders','total_wanters','total_wishers']])
plt.show()
```

## The above facet gives the relation between the variables. We observe a strong positive correlation.
